### The following code queries just portland and displays if it might be raining on 4/16/2024

```python
FUTURE_WEATHER_URL = "https://api.openweathermap.org/data/2.5/forecast?"
CITY = "Portland"
complete_url = FUTURE_WEATHER_URL + "q=" + CITY + "&appid=" + API_KEY

response = json.dumps(requests.get(complete_url).json(), indent=4)
response = requests.get(complete_url)

json_weather = response.json()["list"]

correct_ind = None
resp_range = []
for i in range(len(json_weather)):
  count = 0
  if "2024-04-16" in json_weather[i]["dt_txt"] and correct_ind == None:
    correct_ind = i
    count = i
    resp_range = json_weather[correct_ind:count+1]
  elif "2024-04-16" in json_weather[i]["dt_txt"]:
    count = i
    resp_range = json_weather[correct_ind:count+1]

dict_resp: dict = dict()
for i in range(len(resp_range)):
  dict_resp[i] = resp_range[i]

for i in range(len(dict_resp)):
  if "Rain" in dict_resp[i]["weather"][0]["main"]:
    print("City:", CITY) # can see this in
    print("Next class date predicated rain time:", dict_resp[i]["dt_txt"])
    print("Weather Description:", dict_resp[i]["weather"][0]["main"])
    print("IT WILL RAIN NEXT CLASS!!!!!")
    break
```
